```python
#Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load Train data
train = pd.read_csv("Orion_TrainingData.csv")
```

```python
#Variable Analysis -> Member_id (Not Used for Modelling)
print("% of Missing Values = ",train['member_id'].isnull().sum()/train.shape[0]*100)
print("-----------------------------------------------------------")
print("Duplicate Values = ",train['member_id'].duplicated().sum())
```

```python
#Variable Analysis -> loan_status (Target)
print("% of Missing Values = ",train['loan_status'].isnull().sum()/train.shape[0]*100)
print("-----------------------------------------------------------")
print("% of Defaulters = ",train['loan_status'].sum()/train.shape[0]*100)
print("-----------------------------------------------------------")
print("Distribution: ")
sns.countplot(x='loan_status', data = train, palette = 'hls')
```

```python
#Variable Analysis -> Term (Shows good predictor)
print("% of Missing Values = ",train['term'].isnull().sum()/train.shape[0]*100)
print("-----------------------------------------------------------")
print("% of Defaulters: ",'\n',train.groupby('term')['loan_status'].sum()/train.groupby('term')['member_id'].nunique()*100)
print("-----------------------------------------------------------")
print("Bar Chart: ")
pd.crosstab(train['term'],train['loan_status']).plot(kind = 'bar')
plt.show()
print("-----------------------------------------------------------")
print("Stacked Bar Chart: ")
table = pd.crosstab(train['term'],train['loan_status'])
table.div(table.sum(1).astype(float),axis=0).plot(kind = 'bar', stacked = True)
```
